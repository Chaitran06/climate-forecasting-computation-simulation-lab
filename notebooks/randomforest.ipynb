```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

# Load the data
file_path = "nino-3.4_7924_noaa.txt"
data = pd.read_csv(file_path, delim_whitespace=True, header=None)

# Suppose column 0 = year, column 1 = value
X = data.iloc[:,0].values.reshape(-1,1)   # Independent variable
y = data.iloc[:,1].values                 # Dependent variable

# Create Random Forest Model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

# Prediction
y_pred = model.predict(X)

# Plot
plt.scatter(X, y, label="Actual Data")

# Sort values for smooth plotting
sorted_idx = np.argsort(X.flatten())
plt.plot(X[sorted_idx], y_pred[sorted_idx], color='red', label="Random Forest Prediction")

plt.xlabel("Time")
plt.ylabel("Nino 3.4 Index")
plt.title("Random Forest Regression on Nino 3.4 Data")
plt.legend()
plt.show()
```
